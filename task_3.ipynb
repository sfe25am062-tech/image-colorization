```python
import cv2
import numpy as np

img = cv2.imread("test.jpg")

if img is None:
    print("Error: Image 'test.jpg' could not be loaded. Please ensure the file exists and the path is correct.")
else:
    mask = np.zeros(img.shape[:2], np.uint8)

    mask[50:200,50:200] = 255   # Example region

    img[mask==255] = [255,0,0]

    cv2.imwrite("conditional.jpg", img)
```
