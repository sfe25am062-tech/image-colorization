```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    colored = cv2.applyColorMap(gray, cv2.COLORMAP_JET)

    cv2.imshow("Colorized", colored)

    if cv2.waitKey(1)==27:
        break

cap.release()
cv2.destroyAllWindows()
```
