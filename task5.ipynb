```python
import torch
from torchvision.models.segmentation import deeplabv3_resnet50, DeepLabV3_ResNet50_Weights
from torchvision import transforms
from PIL import Image
import numpy as np
import cv2
import requests
from io import BytesIO

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = deeplabv3_resnet50(weights=DeepLabV3_ResNet50_Weights.DEFAULT).eval().to(device)

# Updated to a new working image URL from Pixabay
image_url = "https://cdn.pixabay.com/photo/2018/08/14/13/23/ocean-3605547_1280.jpg"
response = requests.get(image_url)

# Add checks for successful download and content type
if response.status_code != 200:
    raise requests.exceptions.RequestException(f"Failed to download image from {image_url}. Status code: {response.status_code}")
if 'image' not in response.headers.get('Content-Type', ''):
    raise ValueError(f"Downloaded content from {image_url} is not an image. Content-Type: {response.headers.get('Content-Type')}")

with open("test.jpg", "wb") as f:
    f.write(response.content)
print("Sample image 'test.jpg' downloaded successfully.")

img = Image.open(BytesIO(response.content))
tensor = transforms.ToTensor()(img).unsqueeze(0).to(device)

out = model(tensor)['out'][0]
mask = out.argmax(0).cpu().numpy()

img_np = np.array(img.convert("RGB"))
# Class 15 in COCO dataset corresponds to 'person'. We will color it green.
# Note: This image might not contain 'persons', so the green overlay might not appear.
img_np[mask==15] = [0,255,0]

cv2.imwrite("segmentation.jpg", img_np)
print("Segmentation result saved")
```

```python
import torch
import torchvision.transforms as transforms
import numpy as np
import cv2
from PIL import Image
import requests
from io import BytesIO

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load pretrained segmentation model
model = torch.hub.load('pytorch/vision:v0.10.0',
                       'deeplabv3_resnet101',
                       pretrained=True)
model.to(device)
model.eval()

# Image URL (unga image link podu)
url = "YOUR_IMAGE_URL"
response = requests.get(url)

# Open image
img = Image.open(BytesIO(response.content)).convert("RGB")

# Transform
transform = transforms.Compose([
    transforms.ToTensor()
])

tensor = transform(img).unsqueeze(0).to(device)

# Prediction
with torch.no_grad():
    output = model(tensor)['out'][0]

mask = output.argmax(0).cpu().numpy()

# Convert image to numpy
img_np = np.array(img)

# Class 15 = person (COCO dataset)
img_np[mask == 15] = [0, 255, 0]

# Save result
cv2.imwrite("segmentation.jpg", img_np)

print("Segmentation result saved successfully!")
```
