```python
import cv2
import numpy as np

def historical_tone(image, period):
    if period == "WW1":
        return cv2.applyColorMap(image, cv2.COLORMAP_BONE)
    elif period == "1920s":
        return cv2.applyColorMap(image, cv2.COLORMAP_PINK)
    return image

img = cv2.imread("input.jpg", 0)

# Check if image was loaded successfully
if img is None:
    print("Error: Could not load image 'input.jpg'. Please make sure the file exists and the path is correct.")
else:
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    period = "WW1"
    result = historical_tone(img, period)

    cv2.imwrite("output.jpg", result)
```
